### 第一章 统计学习方法概论

```python
#目标函数
import numpy as np
from scipy.optimize import leastsq
import matplotlib.pyplot as plt

def real_func(x):
    return np.sin(2*np.pi*x)

## 多项式拟合函数
def fit_func(w,x):
    f = np.poly1d(w)
    return f(x)
    
## 定义损失函数
def loss_func(w,x,y):
    return fit_func(w,x) - y

x = np.linspace(0,1,100)
y_=real_func(x) 
#加上正态分布噪音的目标函数
y = [np.random.normal()+y1 for y1 in y_]


x_point = np.linspace(0,1,1000)

def fitting(M=0):
    p_init = np.random.rand(M+1) #选取随机的初始化系数
    p_least = leastsq(loss_func,p_init,args=(x,y)) #用最小二乘法拟合多项式系数
    
    plt.plot(x_point,real_func(x_point),label="real")
    plt.plot(x_point,fit_func(p_least[0],x_point), label="fit")
```

```python
fitting(M=9)
```
